```python
import os
import shutil
import pandas as pd

# --- PATHS ---
source_dir = (
    "/Users/sm6511/Downloads/"
    "test_488998_2026-01-12_22h41.50_e64f68ad-618b-439d-82af-3098e59f55f0/data"
)

dest_dir = (
    "/Users/sm6511/Desktop/Prediction-Accomodation-Exp/"
    "data/Study1.0/Predict"
)
target_date = "2026-01-12"

os.makedirs(dest_dir, exist_ok=True)

# --- COLUMN TO CHECK ---
completion_col = "survey_debrief.block_1/debriefComplete"

moved = []
skipped = []


for fname in os.listdir(source_dir):
    if not fname.endswith(".csv"):
        continue

    if target_date not in fname:   # <-- only process files with this date
        continue

    src_path = os.path.join(source_dir, fname)

    src_path = os.path.join(source_dir, fname)

    try:
        df = pd.read_csv(src_path)
        df.columns = df.columns.str.strip()
    except Exception as e:
        print(f"Could not read {fname}: {e}")
        skipped.append((fname, "read error"))
        continue

    if completion_col not in df.columns:
        print(f"⚠️ Missing completion column in {fname}")
        skipped.append((fname, "missing column"))
        continue

    # Check if any row contains 'complete'
    is_complete = (
        df[completion_col]
        .astype(str)
        .str.lower()
        .eq("complete")
        .any()
    )

    if is_complete:
        dest_path = os.path.join(dest_dir, fname)
        shutil.move(src_path, dest_path)
        #print(f"[DRY RUN] Would move: {fname}")
        moved.append(fname)
        print(f" MOVED: {fname}")
    else:
        skipped.append((fname, "not complete"))

# --- SUMMARY ---
print("\n===== SUMMARY =====")
print(f"Moved files ({len(moved)}):")
for f in moved:
    print(f"  {f}")

print(f"\nSkipped files ({len(skipped)}):")
for f, reason in skipped:
    print(f"  {f} — {reason}")
```
